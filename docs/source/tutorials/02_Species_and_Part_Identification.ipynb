# 02: Fish Species and Part Identification

```python
import os, sys
from pathlib import Path
try:
    import fishy
except ImportError:
    for p in [Path("."), Path(".."), Path("../../..")]:
        if (p / "fishy").exists(): sys.path.append(os.path.abspath(str(p))); break
DATA_PATH = None
for p in [Path("data/REIMS.xlsx"), Path("../data/REIMS.xlsx"), Path("../../../data/REIMS.xlsx")]:
    if p.exists(): DATA_PATH = str(p); break
if DATA_PATH is None: DATA_PATH = "data/REIMS.xlsx"
from fishy import TrainingConfig, run_unified_training, display_final_summary
```

```python
res = run_unified_training(TrainingConfig(model="rf", dataset="species", file_path=DATA_PATH, wandb_log=False))
display_final_summary(res)
```
