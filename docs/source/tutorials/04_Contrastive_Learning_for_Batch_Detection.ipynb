# 04: Contrastive Learning for Batch Detection

```python
import os, sys
from pathlib import Path
try:
    import fishy
except ImportError:
    for p in [Path("."), Path(".."), Path("../../..")]:
        if (p / "fishy").exists(): sys.path.append(os.path.abspath(str(p))); break
DATA_PATH = None
for p in [Path("data/REIMS.xlsx"), Path("../data/REIMS.xlsx"), Path("../../../data/REIMS.xlsx")]:
    if p.exists(): DATA_PATH = str(p); break
if DATA_PATH is None: DATA_PATH = "data/REIMS.xlsx"
from fishy.experiments.contrastive import ContrastiveConfig, run_contrastive_experiment
```

```python
res = run_contrastive_experiment(ContrastiveConfig(dataset="batch-detection", file_path=DATA_PATH, num_epochs=5))
print(f"Accuracy: {res.get('val_accuracy', 0):.4f}")
```
