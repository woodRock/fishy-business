# 03: Oil and Cross-species Adulteration

```python
from fishy import get_data_path, TrainingConfig, run_unified_training, display_final_summary

DATA_PATH = get_data_path()
```

```python
res = run_unified_training(TrainingConfig(model="rf", dataset="oil", file_path=DATA_PATH, wandb_log=False))
import numpy as np; oil_map = {0: 50, 1: 25, 2: 10, 3: 5, 4: 1, 5: 0.1, 6: 0}
preds = res["predictions"]; y_true = np.array([oil_map[idx] for idx in preds["labels"]]); y_pred = np.array([oil_map[idx] for idx in preds["preds"]])
from sklearn.metrics import mean_absolute_error; print(f"MAE: {mean_absolute_error(y_true, y_pred):.2f}%")
```
