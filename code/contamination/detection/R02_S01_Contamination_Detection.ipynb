## Grubbs Test

This method performs a Grubb's test for outlier thresholding
on a marine biomass sample drawn from Mass Spectrometry data.

It is remarkably simple, dumb, elegant solution to the problem.
If the sample contains any outliers, as defined by the Grubb's test,
it is contaminated, otherwise, it is safe.
